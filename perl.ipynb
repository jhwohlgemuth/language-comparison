#### Perl has built-in range operators, regular expressions, string interpolation, and substrings. `List::Util` is a core module that provides `reduce` and `sum`.

```perl
use strict;
use warnings;
use feature 'say';
use List::Util qw(reduce);

my $start = 1;
my $stop  = 100;

sub add {
    my ($x, $y) = @_;
    return $x + $y;
}

my @numbers = $start .. $stop;
my $result  = reduce { add($a, $b) } 0, @numbers;

my $message = sprintf 'The sum of the integers, %d to %d is %d', $start, $stop, $result;
my ($count) = $message =~ /(\d{3})/;
my $answer  = substr $message, -4;

say "Gauss added the first $count numbers and got $answer";
```

### List processing with `grep` and `map`

```perl
use strict;
use warnings;
use feature 'say';

my @over_50_even = grep { $_ > 50 && $_ % 2 == 0 } 1 .. 100;
say '[' . join(',', @over_50_even) . ']';

# Same idea, written as chained filters.
my @same = grep { $_ % 2 == 0 } grep { $_ > 50 } 1 .. 100;
say '[' . join(',', @same) . ']';
```

### Anonymous functions

```perl
use strict;
use warnings;
use feature 'say';

my $double = sub {
    my ($x) = @_;
    return $x * 2;
};

my @doubled = map { $double->($_) } 3, 4, 5;
say '[' . join(',', @doubled) . ']';
```

### Function composition

Perl does not have a built-in function composition operator like Haskell's `.`, but functions are first-class values, so composition can be written as a small helper.

```perl
use strict;
use warnings;
use feature 'say';
use List::Util qw(sum);

sub compose {
    my ($f, $g) = @_;

    return sub {
        return $f->($g->(@_));
    };
}

my $replicate_sum = compose(
    sub {
        my ($xs) = @_;
        return sum @$xs;
    },
    sub {
        my ($x) = @_;
        return [ ($x) x 5 ];
    },
);

say $replicate_sum->(2);
```

### Get primes

```perl
use strict;
use warnings;
use feature 'say';

sub get_primes {
    my ($n) = @_;
    my @primes;

  CANDIDATE:
    for (my $candidate = 2; @primes < $n; $candidate++) {
        for my $prime (@primes) {
            last if $prime * $prime > $candidate;
            next CANDIDATE if $candidate % $prime == 0;
        }

        push @primes, $candidate;
    }

    return @primes;
}

my @first_25 = get_primes(25);
say '[' . join(',', @first_25) . ']';
```
