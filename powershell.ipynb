#### Need to install an external module for FP

```powershell
Install-Module -Name Prelude
```

```powershell
Import-Module -Name Prelude

$Re = '\d{3}'
$Start = 1
$Stop = 100
$Add = { Param($A, $B) $A + $B }
$Numbers = $Start..$Stop
$Result = $Numbers | reduce $Add

$Message = "The sum of the integers, ${Start} to ${Stop}, is ${Result}"
$Count = if ($Message -match $Re) { $Matches[0] } else { '' }
$Answer = $Message.Substring($Message.Length - 4, 4)

"Gauss added the first ${Count} numbers and got ${Answer}"
```
